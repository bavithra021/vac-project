```python
import random
print("clue:programing language")
# List of words to choose from
words = ['python', 'java', 'javascript', 'ruby', 'php', 'csharp', 'html', 'css']

# Function to choose a random word
def choose_word(words):
    return random.choice(words)

# Function to play hangman
def play_hangman():
    word = choose_word(words)
    word_letters = set(word)
    alphabet = set('abcdefghijklmnopqrstuvwxyz')
    used_letters = set()

    lives = 6

    # Loop until the player wins or loses
    while len(word_letters) > 0 and lives > 0:
        # Show the current status
        print('You have', lives, 'lives left and you have used these letters: ', ' '.join(used_letters))

        # Display the word with underscores for missing letters
        word_list = [letter if letter in used_letters else '_' for letter in word]
        print('Current word: ', ' '.join(word_list))

        # Get user input
        user_letter = input('Guess a letter: ').lower()

        # Check if the user's guess is correct
        if user_letter in alphabet - used_letters:
            used_letters.add(user_letter)
            if user_letter in word_letters:
                word_letters.remove(user_letter)
            else:
                lives = lives - 1
                print('Letter is not in the word.')

        # If the user has already guessed that letter
        elif user_letter in used_letters:
            print('You have already used that letter. Guess another letter.')

        # If the user's guess is not a valid letter
        else:
            print('Invalid character. Please guess a letter.')

    # End of game message
    if lives == 0:
        print('You died, sorry. The word was', word)
    else:
        print('Congratulations! You guessed the word', word, '!!')

# Call the play_hangman function to start the game
play_hangman()
```

```python
import tkinter as tk

class TicTacToeGUI:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Tic Tac Toe")

        self.current_player = "X"
        self.board = [[None for _ in range(3)] for _ in range(3)]

        self.buttons = [[None for _ in range(3)] for _ in range(3)]
        for row in range(3):
            for col in range(3):
                self.buttons[row][col] = tk.Button(self.root, text="", font=("Arial", 60), width=2, height=1, command=lambda row=row, col=col: self.button_click(row, col))
                self.buttons[row][col].grid(row=row, column=col)

    def start(self):
        self.root.mainloop()

    def button_click(self, row, col):
        if self.board[row][col] is not None:
            return
        self.board[row][col] = self.current_player
        self.buttons[row][col].config(text=self.current_player)
        if self.check_win():
            tk.messagebox.showinfo("Winner!", f"{self.current_player} wins!")
            self.root.quit()
            return
        if self.check_tie():
            tk.messagebox.showinfo("Tie Game!", "No more moves left.")
            self.root.quit()
            return
        self.current_player = "O" if self.current_player == "X" else "X"

    def check_win(self):
        for i in range(3):
            if self.board[i][0] == self.board[i][1] == self.board[i][2] != None:
                return True
            if self.board[0][i] == self.board[1][i] == self.board[2][i] != None:
                return True
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != None:
            return True
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != None:
            return True
        return False

    def check_tie(self):
        for row in self.board:
            for col in row:
                if col is None:
                    return False
        return True

if __name__ == "__main__":
    gui = TicTacToeGUI()
    gui.start()
```
